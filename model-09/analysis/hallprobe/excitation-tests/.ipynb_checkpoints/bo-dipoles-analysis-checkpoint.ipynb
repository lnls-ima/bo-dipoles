```python
import hallprobe
import matplotlib.pyplot as plt

%matplotlib notebook


currlabels0 = ('0040p31A', '0050p39A', '0060p46A', '0165p27A',
               '0330p54A', '0942p05A', '0991p63A', '1041p21A')
currlabels1 = ('0050p39A', '0991p63A')
currlabels2 = ('0040p31A', '0050p39A', '0991p63A')

magnets = {
    'bd-006': currlabels0,
    'bd-016': currlabels1,
    'bd-027': currlabels2,
    'bd-042': currlabels0,
    'bd-049': currlabels0,
}

data = hallprobe.DipolesSet('x0-9p0450mm-old', magnets)
```

```python
plt.figure()
data.plot_energies(plt)
plt.show()
```

```python
plt.figure()
data.plot_angles(plt)
plt.show()
```

```python
plt.figure()
data.plot_quadrupoles(plt)
plt.show()
```

```python
plt.figure()
data.plot_sextupoles(plt)
plt.show()
```
